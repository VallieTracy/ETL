```python
import pandas as pd
from sqlalchemy import create_engine
```

 # Extract CSVs into DataFrames

```python
# Dataset 1
# Set path for csv file
csvpath = "Resources/Minnesota_Hospitals.csv"
# Import the purchase_data.csv file as a DataFrame
hospitals_df = pd.read_csv(csvpath)
# Preview
hospitals_df.head()
```

```python
# Dataset 2
# Set path for csv file
csvpath = "Resources/indivs_Minnesota18.csv"
# Import file as a DataFrame
mndivs_df = pd.read_csv(csvpath, encoding="ISO-8859-1")

# Preview
mndivs_df.head()
#encoding problem, we were going to try to turn the doc into an excel.  Google
#also wanted to make sure we had the same file names
```

# Transform Hospitals dataframe

```python
# Transform the hospitals_df

# Take out the columns we want
hospital_cols = ["OBJECTID", "NAME", "ADDRESS", "CITY", "STATE", "HOSP_BEDS"]

# Make new copied dataframe so as to not affect the original
hospital_transformed= hospitals_df[hospital_cols].copy()
```

```python
# Clean the data by dropping duplicates and setting the index

#hospital_transformed.drop_duplicates("Hospital ID", inplace=True)
#no duplicates

# Want OBJECTID as our primary key, so set the index to that
hospital_transformed.set_index("OBJECTID", inplace=True)

hospital_transformed
```

# Transform MN Contributions dataframe

```python
mndivs_df.columns
```

```python
# Transform the mndivs_df

# Define columns we want from mndivs_Minnesota18
mndiv_cols = ["Contribid", "Orgname", "Amount", "City", "State", "Zip", "Gender"]

# Transform df with only columns we want, make a copy
mndiv_transformed = mndivs_df[mndiv_cols].copy()

# Rename the column headers
mndiv_transformed = mndiv_transformed.rename(columns={"Contribid": "Contribution_ID"})

# Preview
mndiv_transformed.head()
```

```python
# Clean the data by dropping duplicates and setting the index
mndiv_transformed.drop_duplicates("Contribution_ID", inplace=True)
#no duplicates
mndiv_transformed.set_index("Contribution_ID", inplace=True)
# Preview
mndiv_transformed.head()
```

**Create database connection**

```python
rds_connection_string = "postgres:postgres@localhost:5432/ETL_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Check to see tables created in SQL
engine.table_names()
```

# Load dataframes into database

```python
# SQL Minnesota_Hospitals table
hospital_transformed.to_sql(name="Minnesota_Hospitals", con=engine, if_exists="append", index=False)
```

```python
# SQL indivs_Minnesota18 table
mndiv_transformed.to_sql(name="indivs_Minnesota18", con=engine, if_exists="append", index=False)
```
