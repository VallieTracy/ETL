```python
#set path for csv file
csvpath = "Resources/Minnesota_Hospitals.csv"
#import the purchase_data.csv file as a DataFrame
hospitals_df = pd.read_csv(csvpath)
hospitals_df.head()
```

```python
#set path for csv file
csvpath = "Resources/indivs_Minnesota18.csv"
#import the purchase_data.csv file as a DataFrame
mndivs_df = pd.read_csv(csvpath, encoding="ISO-8859-1")
mndivs_df
#encoding problem, we were going to try to turn the doc into an excel.  Google
#also wanted to make sure we had the same file names
```

```python
# Transform the hospitals_df
hospital_cols = ["OBJECTID", "NAME", "ADDRESS", "CITY", "STATE", "HOSP_BEDS"]
hospital_transformed= hospitals_df[hospital_cols].copy()

# Rename the column headers
hospital_transformed = hospital_transformed.rename(columns={"OBJECTID": "Hospital ID",
                                                          "HOSP_BEDS": "# of Hospital Beds",
                                                          })
hospital_transformed.head()
```

```python
# Clean the data by dropping duplicates and setting the index
#hospital_transformed.drop_duplicates("Hospital ID", inplace=True)
#no duplicates
hospital_transformed.set_index("Hospital ID", inplace=True)

hospital_transformed
```
